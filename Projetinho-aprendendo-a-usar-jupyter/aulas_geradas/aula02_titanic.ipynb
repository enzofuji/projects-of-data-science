```python
import pandas as pd
import numpy as np
teste = pd.read_csv("C:/Users/enzoj/Downloads/UNICAMP/GitHub/projects-of-data-science/Projetinho-aprendendo-a-usar-jupyter/Titanic/test.csv")
train = pd.read_csv("C:/Users/enzoj/Downloads/UNICAMP/GitHub/projects-of-data-science/Projetinho-aprendendo-a-usar-jupyter/Titanic/train.csv")
```

```python
from sklearn.ensemble import RandomForestClassifier #arvores de decisao 

modelo = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
```

```python
#como podemos treinar? entao, devemos primeiro transformar as variaveis mais importantes = mulheres e criancas sao salvas, supostamente, por questao de prioridade

def transformar_sexo(valor):
    if valor == 'female':
        return 1
    else:
        return 0
    
```

```python
train['Sex'].value_counts()
```

```python
train['Sex_binario'] = train['Sex'].map(transformar_sexo) #criei uma coluna nova chamada sex_binario (0 para masculino e 1 para feminino)

train.head()
```

```python
variaveis = ['Sex_binario', 'Age']
X = train[variaveis]
Y = train['Survived']
```

```python
X.head()
```

```python
Y.head()
```

```python
X = X.fillna(-1)
```

```python
modelo.fit(X, Y)
```

```python
X_prev = teste[variaveis]
```

```python
teste['Sex_binario'] = teste['Sex'].map(transformar_sexo)
```

```python
X_prev = teste[variaveis]
X_prev = X_prev.fillna(-1)
X_prev.head()
```

```python

p = modelo.predict(X_prev)
p
```
