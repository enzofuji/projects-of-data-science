```python
dados['country'].unique()
```

```python
dados = dados.drop_duplicates()
dados_usa = dados.query("country == 'USA'")
dados_usa.sort_values("budget", ascending=False).head()
```

```python
budget_gross = dados_usa[["budget", "gross"]].dropna().query("budget >0 | gross > 0")

sns.scatterplot(x="budget", y="gross", data = budget_gross)
```

```python
dados_usa['lucro'] = dados_usa['gross'] - dados_usa['budget'] #lucro/prejuizo

budget_gross = dados_usa.query("budget >0 | gross > 0")[["budget", "lucro"]].dropna()

sns.scatterplot(x="budget", y="lucro", data = budget_gross)
```

```python
budget_gross = dados_usa.query("budget >0 | gross > 0")[["title_year", "lucro"]].dropna()

sns.scatterplot(x="title_year", y="lucro", data = budget_gross)
```

```python
filmes_por_diretor = dados_usa["director_name"].value_counts()
gross_director = dados_usa[["director_name", "gross"]].set_index("director_name").join(filmes_por_diretor, on="director_name")
gross_director.columns=["dindin", "filmes_irmaos"]
gross_director = gross_director.reset_index()
gross_director.head()
```

```python
sns.scatterplot(x="filmes_irmaos", y="dindin", data = gross_director)
```

```python
sns.pairplot(data = dados_usa[["gross", "budget", "lucro", "title_year"]])
```

```python
dados_usa[["gross", "budget", "lucro", "title_year"]].corr()
```
